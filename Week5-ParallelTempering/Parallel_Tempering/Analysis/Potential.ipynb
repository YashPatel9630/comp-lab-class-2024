```python
import numpy as np
import matplotlib.pyplot as plt

# Function to read potential energy data from .xvg file
def read_xvg(file):
    data = []
    with open(file, 'r') as f:
        for line in f:
            if line.startswith('@') or line.startswith('#'):
                continue
            time, energy = map(float, line.split())
            data.append(energy)
    return np.array(data)

# Load potential energy data
energy_T300 = read_xvg('/home/yhp2008/comp-lab-class-2024/Week5-ParallelTempering/Parallel_Tempering/T300/potential_energy_T300.xvg')
energy_T400 = read_xvg('/home/yhp2008/comp-lab-class-2024/Week5-ParallelTempering/Parallel_Tempering/T400/potential_energy_T400.xvg')
energy_T600 = read_xvg('/home/yhp2008/comp-lab-class-2024/Week5-ParallelTempering/Parallel_Tempering/T600/potential_energy_T600.xvg')

# Plot histograms for each temperature
plt.hist(energy_T300, bins=50, alpha=0.5, label='T=300K', density=True)
plt.hist(energy_T400, bins=50, alpha=0.5, label='T=363K', density=True)
plt.hist(energy_T600, bins=50, alpha=0.5, label='T=440K', density=True)

# Add labels and legend
plt.xlabel('Potential Energy (kJ/mol)')
plt.ylabel('Probability Density')
plt.legend()
plt.title('Potential Energy Histograms for Different Temperatures')
plt.show()
```
