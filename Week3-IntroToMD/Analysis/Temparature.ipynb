```python
import numpy as np
import matplotlib.pyplot as plt

# Specify the path to the temperature.xvg file
file_path = '/home/yhp2008/comp-lab-class-2024/Week3-IntroToMD/Setup/temperature.xvg'

# Load the data from temperature.xvg
data = np.loadtxt(file_path, comments=['@', '#'])

# Extract time and temperature values
time = data[:, 0]  # First column (time in ps)
temperature = data[:, 1]  # Second column (temperature in K)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time, temperature, color='red', label='Temperature')
plt.title('Temperature vs. Time')
plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.grid()
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
