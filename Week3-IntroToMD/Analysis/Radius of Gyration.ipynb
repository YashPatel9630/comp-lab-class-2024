```python
import numpy as np
import matplotlib.pyplot as plt

# Function to read the Rg data from the .xvg file
def read_gyrate_xvg(file):
    time, rg = [], []
    with open(file, 'r') as f:
        for line in f:
            if not line.startswith(('@', '#')):
                columns = line.split()
                time.append(float(columns[0]))  # Time (ps)
                rg.append(float(columns[1]))    # Radius of gyration (Rg)
    return np.array(time), np.array(rg)

# Read data from the gyration file
time, rg = read_gyrate_xvg('/home/yhp2008/comp-lab-class-2024/Week3-IntroToMD/Setup/gyrate.xvg')

# Plot the radius of gyration (Rg) over time
plt.figure(figsize=(8, 6))
plt.plot(time, rg, label='Radius of Gyration (Rg)', color='b')
plt.xlabel('Time (ps)', fontsize=14)
plt.ylabel('Rg (nm)', fontsize=14)
plt.title('Radius of Gyration Over Time', fontsize=16)
plt.grid(True)
plt.legend()
plt.show()
```
