```python
import numpy as np
import matplotlib.pyplot as plt

# Define the path to your file
file_path = '/home/yhp2008/comp-lab-class-2024/Week3-IntroToMD/Setup/pressure.xvg'

# Load the data, skipping comments and header lines
data = np.loadtxt(file_path, comments=['#', '@'])

# Extract time and pressure
time = data[:, 0]  # First column is time (ps)
pressure = data[:, 1]  # Second column is pressure (bar)

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(time, pressure, label='Pressure', color='b')
plt.title('Pressure vs Time (NPT Equilibrium)')
plt.xlabel('Time (ps)')
plt.ylabel('Pressure (bar)')
plt.grid()
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
