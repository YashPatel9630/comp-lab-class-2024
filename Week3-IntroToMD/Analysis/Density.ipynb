```python
import numpy as np
import matplotlib.pyplot as plt

# Define the path to your file
file_path = '/home/yhp2008/comp-lab-class-2024/Week3-IntroToMD/Setup/density.xvg'

# Load the data, skipping comments and header lines
data = np.loadtxt(file_path, comments=['#', '@'])

# Extract time and density
time = data[:, 0]  # First column is time (ps)
density = data[:, 1]  # Second column is density (kg/m^3)

# Calculate the 10-ps running average
window_size = 10
running_avg_density = np.convolve(density, np.ones(window_size)/window_size, mode='valid')

# Adjust time array for the running average
time_avg = time[window_size - 1:]  # Time for running average

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(time, density, label='Density', color='b', alpha=0.5)
plt.plot(time_avg, running_avg_density, label='10-ps Running Average', color='r', linewidth=2)
plt.title('Density vs Time (NPT Equilibrium)')
plt.xlabel('Time (ps)')
plt.ylabel('Density (kg/m³)')
plt.grid()
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
