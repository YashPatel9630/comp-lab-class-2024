```python
import numpy as np
import matplotlib.pyplot as plt

# Define the paths to your RMSD files
file_path1 = '/home/yhp2008/comp-lab-class-2024/Week3-IntroToMD/Setup/rmsd.xvg'
file_path2 = '/home/yhp2008/comp-lab-class-2024/Week3-IntroToMD/Setup/rmsd_xtal.xvg'

# Load the first RMSD data, skipping comments and header lines
data1 = np.loadtxt(file_path1, comments=['#', '@'])
time1 = data1[:, 0]  # Time (ns)
rmsd1 = data1[:, 1]  # RMSD (nm)

# Load the second RMSD data, skipping comments and header lines
data2 = np.loadtxt(file_path2, comments=['#', '@'])
time2 = data2[:, 0]  # Time (ns)
rmsd2 = data2[:, 1]  # RMSD (nm)

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(time1, rmsd1, label='RMSD - md_0_1', color='blue', linewidth=2)
plt.plot(time2, rmsd2, label='RMSD - em', color='red', linewidth=2)

plt.title('RMSD vs Time for Backbone')
plt.xlabel('Time (ns)')
plt.ylabel('RMSD (nm)')
plt.grid()
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
