```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the .ener file and load the data
file_path = "/home/yhp2008/comp-lab-class-2024/Week12-IntroAIMD/Proton/proton_aimd_1ns_every0.5fs-1.ener"  # Update the path if needed

# Define column names (as per the data structure in the .ener file)
columns = ["Step", "Time (fs)", "Kin. Energy (a.u.)", "Temp (K)", "Pot. Energy (a.u.)", "Cons Qty (a.u.)", "Used Time (s)"]

# Load the data, skipping the header rows (adjust skiprows if the format changes)
data = pd.read_csv(file_path, delim_whitespace=True, skiprows=1, names=columns)

# Check the first few rows to ensure data is loaded correctly
print(data.head())

# Compute Total Energy (Kinetic + Potential)
data["Total Energy (a.u.)"] = data["Kin. Energy (a.u.)"] + data["Pot. Energy (a.u.)"]

# Plot Potential Energy vs Time
plt.figure(figsize=(10, 6))
plt.plot(data["Time (fs)"], data["Pot. Energy (a.u.)"], label="Potential Energy", color="blue", marker='o', linestyle='-')
plt.xlabel("Time (fs)", fontsize=12)
plt.ylabel("Energy (a.u.)", fontsize=12)
plt.title("Potential Energy vs Time", fontsize=14)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("Potential_Energy_vs_Time.png", dpi=300)  # Save the plot
plt.show()

# Plot Kinetic Energy vs Time
plt.figure(figsize=(10, 6))
plt.plot(data["Time (fs)"], data["Kin. Energy (a.u.)"], label="Kinetic Energy", color="green", marker='o', linestyle='-')
plt.xlabel("Time (fs)", fontsize=12)
plt.ylabel("Energy (a.u.)", fontsize=12)
plt.title("Kinetic Energy vs Time", fontsize=14)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("Kinetic_Energy_vs_Time.png", dpi=300)  # Save the plot
plt.show()

# Plot Total Energy vs Time
plt.figure(figsize=(10, 6))
plt.plot(data["Time (fs)"], data["Total Energy (a.u.)"], label="Total Energy (Kinetic + Potential)", color="red", marker='o', linestyle='-')
plt.xlabel("Time (fs)", fontsize=12)
plt.ylabel("Energy (a.u.)", fontsize=12)
plt.title("Total Energy vs Time", fontsize=14)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("Total_Energy_vs_Time.png", dpi=300)  # Save the plot
plt.show()
```
